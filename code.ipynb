### Serial Recall Function

```python
def serial_recall(length=15, wait=1, n=1, WorkingMemory=0, Pause=0, chunking=False):
    print("\n" * 50)
    time.sleep(3)
    # proportions = []
    # counts = np.zeros(length)
    all_correct_answers = []
    all_user_answers = []
    for i in range(n):

        if chunking:
            choices = ['for',  'fra',  'far',  'mor',  'hun',  'han',  'den',  'det',  'der',  'har',  'som',  'med',  'dig',  'mig',  'sig',  'var',  'vil',  'kan',  'man',  'ned',  'ind',  'til',  'lys',  'lys',  'hus',  'mus',  'kat',  'gød',  'gul',  'rød',  'blå',  'her',  'der',  'vor',  'jeg',  'top',  'arm',  'ben',  'øje',  'fod',  'gød',  'gæs',  'and',  'træ',  'bus',  'bil',  'bog',  'pen',  'dør',  'tag',  'vej','sti',  'sol']
        else:
            choices = [chr(i) for i in range(ord('A'), ord('Z') + 1)]
            
        serial_recall = [random.choice(choices) for i in range(length)]
        all_correct_answers.append(serial_recall.copy())

        for letter in serial_recall:
            print("\n" * 25)
            print(letter)
            print("\n" * 10)
            time.sleep(wait)

        print("\n"*50)

        if WorkingMemory > 0:
            t_end = time.time() + WorkingMemory
            while time.time() < t_end:
                print(t_end - time.time())
                try:
                    inputimeout(f"{random.choice(range(100))} + {random.choice(range(100))} = ", t_end - time.time())
                except TimeoutOccurred:
                    print("\n.............................................................")

        if Pause > 0:
            print(f"Wait for {Pause} seconds")
            time.sleep(Pause)

        time.sleep(1)
        print("You should now enter the sequence:")
        answer = [input("Enter sequence:  ") for letter in serial_recall]
        all_user_answers.append(answer.copy())
        # sum_ = 0
        # for i, x in enumerate(zip(answer,serial_recall)):
        #     if x[0] == x[1]:
        #         sum_ += 1
        #         counts[i] += 1

        # proportion = sum_ / len(answer)
        # proportions.append(proportion)

    # plt.plot(range(1, length + 1), counts)
    # plt.show()
    return np.array(all_correct_answers), np.array(all_user_answers)

```

### Save Data Function

```python
def save_data(corMatrix, recallMatrix,data_file_name_input="data.csv"):

    dataDf = pd.DataFrame(
            np.hstack((corMatrix, recallMatrix)),
            columns=[
                (
                    f"Correct_NumberItem{i}"
                    if i <= corMatrix.shape[1]
                    else f"Answered_NumberItem{i%corMatrix.shape[1]}"
                )

                for i in range(1, corMatrix.shape[1] + recallMatrix.shape[1] + 1)
            ],
        )

    if os.path.isfile(data_file_name_input):
        openDf = pd.read_csv(data_file_name_input, header=None)

        newDataDf = pd.DataFrame(np.vstack((openDf, dataDf)))
        newDataDf.to_csv(data_file_name_input, index=False, header=False)
    else:
        dataDf.to_csv(data_file_name_input, index=False, header=False)
```
